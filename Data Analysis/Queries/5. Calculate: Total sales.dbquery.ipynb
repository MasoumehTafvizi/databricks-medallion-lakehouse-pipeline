```sql
SELECT SUM(sales_amount*price) AS Total_sales, 
    COUNT(DISTINCT customer_ID) AS Total_customers, 
    COUNT(DISTINCT product_id) AS Total_products, 
    COUNT(DISTINCT order_number) 
FROM workspace.salesdb.dim_products p 
    LEFT JOIN workspace.salesdb.fact_sales s ON p.product_key = s.product_key 
    LEFT JOIN workspace.salesdb.dim_customers c ON s.customer_key = c.customer_key



```
