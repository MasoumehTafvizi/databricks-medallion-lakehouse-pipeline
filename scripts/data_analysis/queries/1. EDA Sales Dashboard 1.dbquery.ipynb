```sql
SELECT
    ROUND(SUM(s.sales_amount), 2) AS total_sales,
    ROUND(
        SUM(s.sales_amount - (s.quantity * p.cost)),
        2
    ) AS total_profit,
    SUM(s.quantity) AS total_quantity
FROM workspace.salesdb.fact_sales AS s
LEFT JOIN workspace.salesdb.dim_products AS p
    ON s.product_key = p.product_key;
```
