# The Transformation logic

```python
query = """
SELECT
    ROW_NUMBER() OVER(
        ORDER BY pn.start_date, pn.product_number
    ) AS product_key,
    pn.product_id,
    pn.product_number,
    pn.product_name,
    pc.category_id,
    pc.category,
    pc.sub_category,
    pc.maintenance_flag,
    pn.product_line,
    pn.start_date
FROM silver.crm_products pn
LEFT JOIN silver.erp_product_category pc
    ON pc.sub_category LIKE CONCAT(pn.product_line, '%')
WHERE pn.end_date IS NULL
"""
```

```python
df = spark.sql(query)
display(df)
```

# Wrtting Gold Table

```python
df.write.mode("overwrite").saveAsTable("workspace.gold.dim_products")
```
