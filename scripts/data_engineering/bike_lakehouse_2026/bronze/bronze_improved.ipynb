#Read From CSV and Wite Silver Table

```python
for item in INGESTION_CONFIG:
    # Remove .csv extension from table name if present
    table_name = item['table'].replace('.csv', '')
    full_table_name = f"workspace.bronze.{table_name}"
    
    print(f"Ingestion {item['source']} -> {full_table_name}")

    df = (
        spark.read
            .option("header", True)
            .option("inferSchema", True)
            .csv(f"{item['path']}")
    )
    
    (
        df.write
            .mode("overwrite")
            .format("delta")
            .saveAsTable(full_table_name)
    )
```
